```python
import requests
import json
import math
import time
from datetime import datetime, timedelta
from concurrent.futures import ThreadPoolExecutor, as_completed

# ==========================================
# 1. ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì (Helper Functions)
# ==========================================

def timestamp_to_date(ts_ms):
    if not ts_ms: return "-"
    dt_object = datetime.fromtimestamp(ts_ms / 1000)
    return dt_object.strftime("%d-%m-%Y %H:%M")

def ms_to_duration(ms):
    if not ms: return "00:00:00"
    seconds = int(ms / 1000)
    return str(timedelta(seconds=seconds))

def ms_to_duration_stop(ms):
    if not ms: return "0"
    seconds = ((ms / 1000)/3600)
    return math.ceil(seconds)

def calc_hours_passed(ts_ms):
    if not ts_ms: return 0
    start_time = datetime.fromtimestamp(ts_ms / 1000)
    now = datetime.now()
    diff = now - start_time
    hours_float = diff.total_seconds() / 3600
    return math.ceil(hours_float)

def calculate_completion_time(hours_passed):
    if hours_passed >= 48:
        return "‡πÄ‡∏Å‡∏¥‡∏ô 48 ‡∏ä‡∏°"
    hours_remaining = 48 - hours_passed
    completion_time = datetime.now() + timedelta(hours=hours_remaining)
    return completion_time.strftime('%d/%m/%y %H:%M')

def prepare_headers(cookie_str):
    cookies = {}
    for cookie in cookie_str.split('; '):
        if '=' in cookie:
            key, value = cookie.split('=', 1)
            cookies[key] = value.strip()
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
        'Content-Type': 'application/json',
        'x-csrftoken': cookies.get('csrftoken', ''),
        'Referer': 'https://creator.shopee.co.th/insight/live/list',
    }
    return headers, cookies

# ==========================================
# 2. ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• (Worker Function)
# ==========================================
# ‡∏õ‡∏£‡∏±‡∏ö‡πÅ‡∏Å‡πâ: ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤ name, cookie ‡πÄ‡∏Ç‡πâ‡∏≤‡∏°‡∏≤‡∏ï‡∏£‡∏á‡πÜ ‡πÅ‡∏•‡∏∞ Return ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Å‡∏•‡∏±‡∏ö‡πÑ‡∏õ (‡πÑ‡∏°‡πà‡πÅ‡∏Å‡πâ Global ‡πÄ‡∏≠‡∏á)

def fetch_shopee_data(r_name, r_cookie):
    
    # URL
    API_URL_user = "https://creator.shopee.co.th/supply/api/lm/sellercenter/userInfo"
    API_URL_live = "https://creator.shopee.co.th/supply/api/lm/sellercenter/realtime/sessionList"
    
    headers, cookies = prepare_headers(r_cookie)
    params = {"page": 1, "pageSize": 100}
    
    results = [] # ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ï‡∏∞‡∏Å‡∏£‡πâ‡∏≤‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏ô‡∏ô‡∏µ‡πâ
    
    try:
        # 1. ‡πÄ‡∏ä‡πá‡∏Ñ‡∏ä‡∏∑‡πà‡∏≠ User (UserInfo)
        # ------------------------------------------
        resp_user = requests.get(API_URL_user, cookies=cookies, headers=headers, timeout=10)
        if resp_user.status_code != 200: return []
        
        data_user = resp_user.json()
        ck_name = data_user.get('data', {})
        fetched_name = ck_name.get('userName', '') or ck_name.get('name', '')

        if fetched_name != r_name:
            # print(f"‡∏ä‡∏∑‡πà‡∏≠‡πÑ‡∏°‡πà‡∏ï‡∏£‡∏á: ‡πÄ‡∏õ‡πâ‡∏≤‡∏´‡∏°‡∏≤‡∏¢ {r_name} != ‡∏ó‡∏µ‡πà‡πÑ‡∏î‡πâ {fetched_name}")
            # ‡∏≠‡∏≤‡∏à‡∏à‡∏∞ return [] ‡∏´‡∏£‡∏∑‡∏≠‡∏à‡∏∞‡πÄ‡∏Å‡πá‡∏ö error log ‡∏Å‡πá‡πÑ‡∏î‡πâ
            return []

        # 2. ‡∏î‡∏∂‡∏á‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£ Live (SessionList)
        # ------------------------------------------
        resp_live = requests.get(API_URL_live, cookies=cookies, headers=headers, params=params, timeout=10)
        if resp_live.status_code != 200: return []
        
        data_live = resp_live.json()
        live_list = data_live.get('data', {}).get('list', [])
        
        # ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏°‡∏µ session ‡πÄ‡∏•‡∏¢ ‡πÉ‡∏´‡πâ return ‡∏Ñ‡πà‡∏≤ default (‡∏Å‡∏£‡∏ì‡∏µ‡∏™‡∏ñ‡∏≤‡∏ô‡∏∞ "‡πÑ‡∏•‡∏û‡πå" ‡πÅ‡∏ï‡πà‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡πÑ‡∏î‡πâ‡πÄ‡∏£‡∏¥‡πà‡∏°)
        if not live_list:
            results.append({
                "sessionId": 0,
                "username": r_name,
                "placedSales": 0,
                "status": "WAITING",
                "time_info": 0,
                "end_time_info": "-",
                "placedOrders": 0,
                "productClicks": 0,
                "atc": 0,
                "startTime": "-",
                "product": 0
            })
            return results
        
        # ‡∏ß‡∏ô‡∏•‡∏π‡∏õ‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏Ç‡πâ‡∏≤‡∏ï‡∏∞‡∏Å‡∏£‡πâ‡∏≤ results
        for item in live_list:
            
            # ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡πà‡∏≤‡∏ï‡πà‡∏≤‡∏á‡πÜ
            duration = item.get('duration', 0)
            start_time_raw = item.get('startTime')
            #hour_passed = calc_hours_passed(start_time_raw)
            
            status_text = "LIVE" if duration == 0 else "OFFLINE"
            
            if duration == 0: # ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡πÑ‡∏•‡∏ü‡πå
                 hour_passed = calc_hours_passed(start_time_raw)
                 end_time_txt = "‡πÄ‡∏Å‡∏¥‡∏ô 48 ‡∏ä‡∏°" if hour_passed > 48 else calculate_completion_time(hour_passed)
                 time_display = hour_passed
            else: # ‡∏à‡∏ö‡πÑ‡∏•‡∏ü‡πå‡πÅ‡∏•‡πâ‡∏ß
                 time_display = ms_to_duration_stop(duration)
                 # ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡πÄ‡∏ß‡∏•‡∏≤‡∏à‡∏ö‡∏à‡∏£‡∏¥‡∏á = start + duration
                 end_time_dt = datetime.fromtimestamp((start_time_raw + duration) / 1000)
                 end_time_txt = end_time_dt.strftime("%d-%m-%Y %H:%M")

            #3. ‡∏î‡∏∂‡∏á‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£ ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤
            # ------------------------------------------
            
            try:
                API_URL_product   = f"https://creator.shopee.co.th/supply/api/lm/sellercenter/realtime/dashboard/productList?sessionId={item.get('sessionId', 0)}&productName=&productListTimeRange=0&productListOrderBy=productClicks&sort=desc&page=1&pageSize=100"
                resp_product = requests.get(API_URL_product, cookies=cookies, headers=headers, params=params, timeout=10)
                #if resp_product.status_code != 200: return []
                
                data_product = resp_product.json()
                product_list = data_product.get('data', {}).get('list', [])
                product_lan = len(product_list)
            except:product_lan = "xx"


            # ‡πÄ‡∏Å‡πá‡∏ö‡πÉ‡∏™‡πà Dictionary (‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡πÉ‡∏ä‡πâ‡∏á‡πà‡∏≤‡∏¢‡πÅ‡∏•‡∏∞‡πÄ‡∏õ‡πá‡∏ô‡∏£‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ö)
            row_data = {
                "sessionId": item.get('sessionId', 0),
                "username": r_name,
                # ‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç 4 ‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡∏î‡πâ‡∏≤‡∏ô‡∏•‡πà‡∏≤‡∏á‡∏ô‡∏µ‡πâ ‡πÇ‡∏î‡∏¢‡πÄ‡∏û‡∏¥‡πà‡∏° ( ... or 0)
                "placedSales": int(float(item.get('placedSales') or 0)), 
                "status": status_text,
                "time_info": time_display, 
                "end_time_info": end_time_txt,
                "placedOrders": int(float(item.get('placedOrders') or 0)),
                "productClicks": int(float(item.get('productClicks') or 0)),
                "atc": int(float(item.get('atc') or 0)),
                "startTime": timestamp_to_date(start_time_raw),
                "product": product_lan
            }
            
            results.append(row_data)
            break
            
        # *** ‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç‡∏à‡∏∏‡∏î‡∏ú‡∏¥‡∏î: ‡∏¢‡πâ‡∏≤‡∏¢ return ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤‡∏ô‡∏≠‡∏Å for loop ***
        # ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Ñ‡∏£‡∏ö‡∏ó‡∏∏‡∏Å‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡πÉ‡∏ô page ‡∏ô‡∏±‡πâ‡∏ô ‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πà‡πÅ‡∏Ñ‡πà‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡πÅ‡∏£‡∏Å
        return results

    except Exception as e:
        print(f"Error fetching {r_name}: {e}")
        return []
```

```python

shopee_live = {}
loop_work = 10 #‡πÄ‡∏≠‡∏≤‡πÑ‡∏ß‡πâ‡∏ï‡∏±‡πâ‡∏á‡∏ß‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏£‡∏≠‡∏ö‡∏à‡∏∞‡πÑ‡∏õ‡∏ó‡∏≥‡∏á‡∏≤‡∏ô‡∏™‡∏ñ‡∏≤‡∏ô‡∏∞‡∏≠‡∏∑‡πà‡∏ô
def main_process():
    global shopee_live,data_dict,loop_work

    start_time = time.time()

    # ‡πÉ‡∏ä‡πâ ThreadPoolExecutor ‡∏¢‡∏¥‡∏á‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏Å‡∏±‡∏ô 20 ‡∏á‡∏≤‡∏ô (Workers)
    # ‡∏õ‡∏£‡∏±‡∏ö‡πÄ‡∏•‡∏Ç max_workers ‡πÑ‡∏î‡πâ (‡πÅ‡∏ô‡∏∞‡∏ô‡∏≥ 10-20 ‡∏ñ‡πâ‡∏≤‡πÄ‡∏¢‡∏≠‡∏∞‡πÑ‡∏õ Shopee ‡∏≠‡∏≤‡∏à‡∏ö‡∏•‡πá‡∏≠‡∏Å)
    with ThreadPoolExecutor(max_workers=20) as executor:
        futures = []
        
        # 1. ‡∏™‡∏±‡πà‡∏á‡∏á‡∏≤‡∏ô (Submit Tasks)
        print(f"üöÄ ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• {len(data_dict)} ‡∏£‡πâ‡∏≤‡∏ô‡∏Ñ‡πâ‡∏≤...")
        for name in data_dict:
            # ‡πÄ‡∏ä‡πá‡∏Ñ‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç‡∏Å‡πà‡∏≠‡∏ô‡∏¢‡∏¥‡∏á

            if loop_work > 4 :
                if data_dict[name]["status"] == "‡πÑ‡∏•‡∏û‡πå" or data_dict[name]["status"] == "‡∏ö‡∏±‡∏ç‡∏ä‡∏µ‡∏û‡∏£‡πâ‡∏≠‡∏°" or data_dict[name]["status"] == "new ‡πÑ‡∏°‡πà‡∏ï‡∏¥‡∏î‡∏à‡∏≠" or data_dict[name]["status"] == "old ‡πÑ‡∏°‡πà‡∏ï‡∏¥‡∏î‡∏à‡∏≠" or data_dict[name]["status"] == "‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏ó‡∏î‡∏™‡∏≠‡∏ö ‡∏à‡∏≠":
                    futures.append(executor.submit(fetch_shopee_data, name, data_dict[name]["cookie"]))
            else:
                if data_dict[name]["status"] == "‡πÑ‡∏•‡∏û‡πå":
                    futures.append(executor.submit(fetch_shopee_data, name, data_dict[name]["cookie"]))
        if loop_work > 4 : loop_work = 0
        else:loop_work += 1

        # 2. ‡∏£‡∏≠‡∏£‡∏±‡∏ö‡∏á‡∏≤‡∏ô‡πÅ‡∏•‡∏∞‡πÄ‡∏Å‡πá‡∏ö‡∏ú‡∏• (Collect Results)
        print(f"üì¶ ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•... (‡∏¢‡∏¥‡∏á‡πÑ‡∏õ {len(futures)} Requests)")
        
        for future in as_completed(futures):
            data_rows = future.result() # ‡∏£‡∏±‡∏ö‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡∏ó‡∏µ‡πà return ‡∏°‡∏≤ (‡πÄ‡∏õ‡πá‡∏ô list ‡∏Ç‡∏≠‡∏á dict)
            
            if data_rows:
                # ‡∏ß‡∏ô‡∏•‡∏π‡∏õ‡πÅ‡∏Å‡∏∞‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏™‡πà Global List ‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏∏‡∏ì
                for row in data_rows:
                    name = row['username']
                    try :shopee_live[name]
                    except:shopee_live[name] = {} #‡∏õ‡πâ‡∏≠‡∏á‡∏Å‡∏±‡∏ô‡∏Å‡∏≤‡∏£‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ã‡πâ‡∏≥‡πÅ‡∏•‡πâ‡∏ß‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏≤‡∏¢
                    shopee_live[name]['sessionId'] = row['sessionId']
                    shopee_live[name]['username'] = row['username']
                    shopee_live[name]['placedSales'] = row['placedSales']
                    shopee_live[name]['status'] = row['status']
                    shopee_live[name]['time_info'] = row['time_info']
                    shopee_live[name]['end_time_info'] = row['end_time_info']
                    shopee_live[name]['placedOrders'] = row['placedOrders']
                    shopee_live[name]['productClicks'] = row['productClicks']
                    shopee_live[name]['atc'] = row['atc']
                    shopee_live[name]['product'] = row['product']
                    shopee_live[name]['starttime'] = row['startTime']
    end_time = time.time()
    duration = end_time - start_time
    print(f"‚úÖ ‡πÄ‡∏™‡∏£‡πá‡∏à‡∏™‡∏¥‡πâ‡∏ô! ‡πÉ‡∏ä‡πâ‡πÄ‡∏ß‡∏•‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î {duration:.2f} ‡∏ß‡∏¥‡∏ô‡∏≤‡∏ó‡∏µ")
    #print(f"üìä ‡πÑ‡∏î‡πâ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î {len(data_rows)} ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£")
```
